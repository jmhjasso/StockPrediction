```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Instruction to upload a csv file, CRMT in this case
from google.colab import files
data_to_load = files.upload()
```

```python
#To show the full file
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
import io
adf = pd.read_csv(io.BytesIO(data_to_load['CRMT.csv']))
adf
```

```python
#removing the not needed columns
adf = adf.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)
adf.head(10)
```

```python
adf.info()
```

```python
#Transofrming the date column from object to datetime
adf['Date'] = pd.to_datetime(adf['Date'])
```

```python
adf.info()
```

```python
#Getting the avg of the close column by month
adf=adf.groupby(pd.Grouper(key='Date', freq='M'))['Close'].mean() #M is for monthly
adf
```

```python
#Instruction to upload unemployment csv file
from google.colab import files
data_to_load = files.upload()
```

```python
#reading unemployment rates csv and assigning it to a variable
import io
adu = pd.read_csv(io.BytesIO(data_to_load['Unemployment rates.csv']))
adu
```

```python
adu.info()
```

```python
#Transforming the date column from object to datetime
adu['Date'] = pd.to_datetime(adu['Date'])
```

```python
adu.head(5)
```

```python
#appending stock info with unemployment info
data_merged = pd.merge(adu, adf, on='Date')
data_merged
```

```python
#Instruction to upload inflation ratescsv file
from google.colab import files
data_to_load = files.upload()
```

```python
#reading inflation rates csv and assigning it to a variable
import io
adi = pd.read_csv(io.BytesIO(data_to_load['Inflation.csv']))
adi
```

```python
#removing the not needed columns
adi = adi.drop(['Unnamed: 2', 'Unnamed: 3'], axis=1)
adi.head(10)
```

```python
#Transforming the date column from object to datetime
adi['Date'] = pd.to_datetime(adi['Date'])
```

```python
#appending stock info with unemployment info
data_merged = pd.merge(data_merged, adi, on='Date')
data_merged
```

```python
#changing the name of some columns
data_merged.rename(columns={'Total': 'Unemployment Rate'}, inplace=True)
data_merged.rename(columns={'Close': 'Stock Price'}, inplace=True)
data_merged
```

```python
data_merged.to_csv('CRMT and economic rates.csv', index=False)
```
